```python
plt.scatter(M.radius_mean, M.texture_mean, color="red", label="M",alpha=0.3)
plt.scatter(B.radius_mean, B.texture_mean, color="green", label="B",alpha=0.3)
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
plt.legend()
plt.show()
```

```python
df.diagnosis= [1 if each== "M" else 0 for each in df.diagnosis]
y= df.diagnosis.values
x_df= df.drop(["diagnosis"],axis=1)
```

```python
#normalization
x= (x_df - np.min(x_df))/(np.max(x_df)-np.min(x_df))
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test=train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
#naive-bayes model

from sklearn.naive_bayes import GaussianNB
nb= GaussianNB()
nb.fit(x_train,y_train)
```

```python
print( "{} naive bayes score : {}".format(3,nb.score(x_test,y_test)))
```
